```python
# Pregunta 1:
import yfinance as yf
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='1y')
tesla_data.reset_index(inplace=True)
print("PREGUNTA 1: Datos de acciones de Tesla")
print(tesla_data.head())
```

```python
# PREGUNTA 2:
import pandas as pd
import requests
url_tesla = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url_tesla, headers=headers)
all_tables = pd.read_html(response.text)
tesla_revenue = None
for table in all_tables:
    if table.shape[1] == 2 and 'Tesla Quarterly Revenue' not in table.columns:
        tesla_revenue = table
        break
if tesla_revenue is not None:
    tesla_revenue.columns = ['Fecha', 'Ingresos']
    tesla_revenue.dropna(inplace=True)
    tesla_revenue['Ingresos'] = tesla_revenue['Ingresos'].replace('[\$,]', '', regex=True).astype(float)
    print("PREGUNTA 2: Últimas 5 filas del DataFrame tesla_revenue")
    print(tesla_revenue.tail())
else:
    print("No se encontró la tabla de ingresos de Tesla.")
```

```python
# PREGUNTA 3:
import yfinance as yf
gme = yf.Ticker('GME')
gme_data = gme.history(period='1y')
gme_data.reset_index(inplace=True)
print("PREGUNTA 3: Datos de acciones de GameStop")
print(gme_data.head())
```

```python
# PREGUNTA 4:
import pandas as pd
import requests
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
try:
    tables = pd.read_html(response.text)
except ValueError:
    print("❌ No se encontraron tablas en la página.")
    tables = []
gme_revenue = None
for table in tables:
    if table.shape[1] == 2 and 'Revenue' not in table.columns[0]:
        gme_revenue = table
        break
if gme_revenue is not None:
    gme_revenue.columns = ['Fecha', 'Ingresos']
    gme_revenue.dropna(inplace=True)
    gme_revenue['Ingresos'] = gme_revenue['Ingresos'].replace('[\$,]', '', regex=True).astype(float)
    print("PREGUNTA 4: Últimas 5 filas del DataFrame gme_revenue")
    print(gme_revenue.tail())
else:
    print("No se encontró una tabla de ingresos válida para GameStop.")
```

```python
# PREGUNTA 5:
import yfinance as yf
import matplotlib.pyplot as plt
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='1y')
tesla_data.reset_index(inplace=True)
def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data['Date'], data['Close'], linewidth=2)
    plt.title(title)
    plt.xlabel("Fecha")
    plt.ylabel("Precio de Cierre (USD)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
make_graph(tesla_data, "PREGUNTA 5: Precio de Cierre de Tesla - Último Año")
```
